```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_thread = pd.read_csv("stress_test_results_thread.csv")
data_process = pd.read_csv("stress_test_results_process.csv")
```

```python
data_thread
```

```python
data_process
```

```python
data =  pd.concat([data_process, data_thread])
```

```python
file_path = 'combined_output.csv'
data.to_csv(file_path, index=False)
```

```python
data
```

```python
df = data
```

```python
# Drop rows where operation is 'list'
df = df[df['operation'] != 'list']
```

```python
# Ganti 'process' dengan 'thread' untuk grafik lainnya
selected_executor = 'process'

# Filter berdasarkan executor_type
df_filtered = df[df['executor_type'] == selected_executor]

# Tambahkan kolom throughput rata-rata manual
df_filtered['mean_throughput'] = df_filtered[
    ['avg_throughput', 'median_throughput', 'min_throughput', 'max_throughput']
].mean(axis=1)

# Group by file_size_mb dan operation
summary = df_filtered.groupby(['file_size_mb', 'operation'])['mean_throughput'].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
for op in summary['operation'].unique():
    data_op = summary[summary['operation'] == op]
    plt.plot(data_op['file_size_mb'], data_op['mean_throughput'], marker='o', label=op)

plt.title(f'Average Throughput by File Size and Operation\nExecutor: {selected_executor}')
plt.xlabel('File Size (MB)')
plt.ylabel('Average Throughput')
plt.grid(True)
plt.legend(title='Operation')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# Hitung success rate
df['success_rate'] = df['success_count'] / (df['success_count'] + df['fail_count'])

# Group by operation dan executor_type
success_summary = df.groupby(['operation', 'executor_type'])['success_rate'].mean().reset_index()

# Visualisasi
plt.figure(figsize=(8, 5))
sns.barplot(data=success_summary, x='operation', y='success_rate', hue='executor_type')

plt.title('Success Rate by Operation and Executor Type')
plt.ylabel('Success Rate')
plt.ylim(0, 1.5)
plt.xlabel('Operation')
plt.legend(title='Executor Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Group by client_pool_size and compute average throughput in MB/s
throughput_by_pool = df.groupby('client_pool_size').apply(
    lambda g: g['file_size_mb'].sum() / g['avg_duration'].sum()
).reset_index(name='average_throughput_MBps')

print(throughput_by_pool)
```

```python
# Group by client_pool_size and compute average throughput in MB/s
throughput_by_pool = df.groupby('server_pool_size').apply(
    lambda g: g['file_size_mb'].sum() / g['avg_duration'].sum()
).reset_index(name='average_throughput_MBps')

print(throughput_by_pool)
```

```python
throughput_by_executor = df.groupby('executor_type').apply(
    lambda g: g['file_size_mb'].sum() / g['avg_duration'].sum()
).reset_index(name='average_throughput_MBps')

print(throughput_by_executor)
```

```python
failures_by_operation = df.groupby('operation')['fail_count'].sum().reset_index(name='total_failures')
print(failures_by_operation)
```
